```python
import pandas as pd

rivers = pd.read_csv('rivers_data_file.csv', sep=';')
rivers.head(50)
```

```python
relevant_rivers = rivers[
    (rivers['Name'] == 'Braid Burn (Upstream Dreghorn Barracks to Portobello)') |
    (rivers['Name'] == 'Water of Leith (Poets Burn to Murray Burn confluences)') |
    (rivers['Name'] == 'Water of Leith (Poet\'s Burn to Murray Burn confluences)')
]

recent_data = relevant_rivers.drop(columns=['Source', 'Location', 'On water body', '2018', '2017', '2016', '2015', '2014', '2013', '2012', '2011', '2010', '2009', '2008', '2007','2020.1', '2019.1', '2018.1', '2017.1', '2016.1', '2015.1', '2014.1', '2013.1', '2012.1', '2011.1', '2010.1', '2009.1', '2008.1', '2007.1'])


display(recent_data)
```

```python
columns_to_keep = ['ID', 'Name', 'Parameter', '2022', '2022.1', '2020', '2019']

subset_recent_data = recent_data[columns_to_keep]

pivoted_recent_data = subset_recent_data.pivot_table(index=['ID', 'Name', 'Parameter'], aggfunc='first').unstack(level=-1)

pivoted_recent_data.reset_index(inplace=True)

display(pivoted_recent_data)
```

```python
columns_to_keep = ['ID', 'Name', 'Parameter', '2022', '2022.1', '2020', '2019']

subset_recent_data = recent_data[columns_to_keep]

pivoted_recent_data = subset_recent_data.pivot_table(index=['ID', 'Name', 'Parameter'], aggfunc='first').unstack(level=-1)

pivoted_recent_data.reset_index(inplace=True)

year_dfs = {} 

for year in pivoted_recent_data.columns.get_level_values(0).unique():
    year_recent_data = pivoted_recent_data[year]
    year_dfs[year] = year_recent_data


available_years = list(year_dfs.keys())
print("Available years:", available_years)


display(year_dfs['2022'])
display(year_dfs['2020'])
display(year_dfs['2019'])

```

```python
for year in available_years:
    year_dfs[year].to_excel(f'data_{year}.xlsx', index=True)
```
